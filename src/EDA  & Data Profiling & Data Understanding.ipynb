##### Frequency of total order purchased by each customer

```python
orders.order_number.describe().apply(lambda x: format(x, '.2f'))

```

```python
order_number_per_customer = orders.groupby('user_id')['order_number'].max()
order_number_per_customer = order_number_per_customer.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax = sns.barplot(x = order_number_per_customer.index, y = order_number_per_customer.values)
ax.set_xlabel('Orders per customer')
ax.set_ylabel('Count')
ax.xaxis.set_tick_params(rotation=90, labelsize=10)
ax.set_title('Frequency of Total Orders by Customers')
fig.savefig('Frequency of Total Order purchased by Customers.png')
```

##### Total Orders per day of a week

```python
fig, ax = plt.subplots(figsize = (15,4))
ordersDay = orders[["order_dow"]].replace({0:"Saturday",1:"Sunday",2:"Monday",3:"Tuesday",4:"Wednesday",5:"Thursday",6:"Friday"})
ax = sns.countplot(data=ordersDay,x="order_dow",ax=ax,order=["Saturday","Sunday","Monday","Tuesday","Wednesday","Thursday","Friday"])
ax.set_xlabel('Day of Week', size = 10)
ax.set_ylabel('Orders', size = 10)
ax.tick_params(axis = 'both', labelsize = 8)
ax.set_title('Total Orders per Day of Week')
fig.savefig('Total Orders per Day of a Week.png')
plt.show()
```

##### Total Unique Users per Day of a Week

```python
order_per_customer_per_day = orders.groupby('order_dow')['user_id'].nunique()
fig, ax = plt.subplots(figsize = (10,3))
ax = sns.barplot(x = order_per_customer_per_day.index, y = order_per_customer_per_day.values)
ax.set_xlabel('Day of Week', size = 10)
ax.set_ylabel('Total Unique Users', size = 10)
ax.tick_params(axis = 'both', labelsize = 8)
ax.set_title('Total Unique Users per Day of Week')
fig.savefig('Total Unique Users per Day of a Week.png')
plt.show()
```

##### Total Orders per hour of day

```python
fig, ax = plt.subplots(figsize = (10,5))
ax = sns.countplot(orders.order_hour_of_day)
ax.set_xlabel('Hour of Day', size = 10 )
ax.set_ylabel('Orders', size = 10)
ax.tick_params(axis = 'both', labelsize = 8)
ax.set_title('Total Orders per Hour of Day')
fig.savefig('Total Orders per Hour of Day.png')
plt.show()
```

##### Orders VS Days since prior order

##### Order Frequency of Day in a week Vs Hour of day

```python
order_hour_of_day_analysis = orders.groupby(["order_dow", "order_hour_of_day"])["order_number"].aggregate("count").reset_index()
order_hour_of_day_analysis = order_hour_of_day_analysis.pivot('order_dow', 'order_hour_of_day', 'order_number')
order_hour_of_day_analysis.head()
```

```python
ax = plt.subplots(figsize=(12,5))
ax = sns.heatmap(order_hour_of_day_analysis, cmap="YlGnBu", linewidths=.5)
ax.set_title("Frequency of Day of week Vs Hour of day", size = 12)
ax.set_xlabel("Hour of Day", size = 10)
ax.set_ylabel("Day of Week", size = 10)
ax.tick_params(axis = 'both', labelsize = 8)
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=10)
fig = ax.get_figure()
fig.savefig("Order Frequency of Day in a week Vs Hour of day.png")
plt.show()
```

##### Relationship between products and Aisle (No of product in each aisle)

```python
total_product_in_aisle = products.groupby('aisle_id')['product_id'].count()

fig, ax = plt.subplots(figsize = (15,6))
ax = sns.barplot(x = total_product_in_aisle.index, y = total_product_in_aisle.values)
ax.set_xlabel('Aisle Id')
ax.set_ylabel('Total products in aisle')
ax.xaxis.set_tick_params(rotation=90, labelsize = 7)
ax.set_title('Total Products in Aisle VS Aisle ID', size = 12)
fig.savefig('Total Products in Aisle VS Aisle ID.png')
```

##### Relationship between products and department (No of product in each department)

```python
total_product_in_department = products.groupby('department_id')['product_id'].count()
fig, ax = plt.subplots(figsize = (8,5))
ax = sns.barplot(x = total_product_in_department.index, y = total_product_in_department.values)
ax.set_xlabel('Department Id')
ax.set_ylabel('Total products in department')
ax.xaxis.set_tick_params(rotation=90, labelsize = 9)
ax.set_title('Total Products in Department VS Department ID', size = 10)
fig.savefig('Total Products in Department VS Department ID.png')
```

##### Relationship between department and Aisle (No of aisle in each department)

```python
total_aisle_in_department = products.groupby('department_id')['aisle_id'].nunique()

fig, ax = plt.subplots(figsize = (8,5))
ax = sns.barplot(x = total_aisle_in_department.index, y = total_aisle_in_department.values)
ax.set_xlabel('Department Id')
ax.set_ylabel('Total Aisles in department')
ax.xaxis.set_tick_params(rotation=90, labelsize = 9)
ax.set_title('Total Aisles in Department VS Department ID', size = 10)
fig.savefig('Total Aisles in Department VS Department ID.png')
```
